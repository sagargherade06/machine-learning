```python
import pandas as pd
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
```

```python
#example data
data={'unemployment rate':[5.5,6.0,7.0,8.5,6.3],
     'gdp':[45000,42000,40000,35000,41000]}
df=pd.DataFrame(data)
```

```python
print(df)
```

```python
x=df[['unemployment rate']]
y=df['gdp']
```

```python
#initialisation model
model=LinearRegression()
model.fit(x,y)
```

```python
#get the coefficient  and intercept
print(f"intercept:{model.intercept_}")
print(f"coeficent and slop:{model.coef_[0]}")
```

```python
prediction=model.predict(x)
```

```python
plt.scatter(x,y,color='blue',label='datapoints')
plt.plot(x,prediction,color='red',label='regressionline')
plt.xlabel('unemployment rate')
plt.ylabel('gdp')
plt.legend()
plt.show()
```
