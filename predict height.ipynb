```python
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
```

```python
import matplotlib.pyplot as plt
```

```python
#data heiht vs weight
height=np.array([150,160,170,180,190]).reshape(-1,1)
weight=np.array([50,55,60,65,70])
```

```python
#create the linear regression model
model=LinearRegression()
```

```python
#fit the model to the data
model.fit(height,weight)
```

```python
print(f"Intercept: {model.intercept_}")
print(f"Slope: {model.coef_[0]}")
```

```python
#predict weight for a given hegiht
prediction_weight=model.predict(np.array([[175]]))
print(f"predicted weight for height 175cm:{prediction_weight[0]}")
```

```python
#ploting the data and the regression line
plt.scatter(height,weight,color='blue',label='data points')
plt.plot(height,model.predict(height),color='red',label='Regression line')
plt.xlabel('height(cm)')
plt.ylabel('weight(kg)')
plt.legend()
plt.show()
```
